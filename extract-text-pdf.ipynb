```python
import PyPDF2
```

```python
f = open('marco-civil-internet.pdf', 'rb')
pdf = PyPDF2.PdfFileReader(f)
for i in range(pdf.getNumPages()):
    print("--- Pagina %d -----------------------------"%i)
    pg = pdf.getPage(i).extractText()
    print(pg)
    print("-------------------------------------------")
 
```
